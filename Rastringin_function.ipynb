```python
# Problem 2b 
#Import
import matplotlib.pyplot as plt
from math import cos, pi, sin
import numpy as np

#function
def rastringin(x,y):
    return (1-cos(pi*x/2))*(y**2)


xi, yi = np.meshgrid( np.linspace(-1, 1, 500), np.linspace(-1, 1, 500) )

f_eval = np.zeros(xi.shape)
mask = []
for i in range(500):
    for j in range(500):
        f_eval[i, j] = rastringin(xi[i,j], yi[i,j])

#plot
plt.figure()
plt.contourf(xi, yi, f_eval)
plt.colorbar()

plt.show()
```

```python
#Problem 2c&2e
def grad_Rastringin(xi, yi):
    return np.array([ (pi*(yi**2)*sin(pi*xi/2))/2, 2*yi*(1-cos(pi*xi/2))])
plt.figure()
plt.axis('equal') 
im = plt.contourf(xi, yi, f_eval, levels = 10)
plt.title("Rastringin function, 10 steps gradient descent")
for j in range(1000): #no need for this for question c
    p = [tuple(np.random.uniform(-1, 1, 2))]
    for i in range(10):
        p.append(tuple(np.subtract(p[-1], grad_Rastringin(*p[-1]))))
    p = np.array(p)
    plt.plot(*p.T, c='k', linewidth=0.5, markersize=1, marker = '.')
plt.show()
```
